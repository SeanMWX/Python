```python
import os,sys
```

```python
def rename_files(path):
    old_names = os.listdir(path)
    for old_name in old_names:
        if not old_name.endswith('.jpg'):
            continue
        num = int(old_name[2:-4])
        if old_name[0] == 'a':
            num = (num - 2) * 2 + 1
        elif old_name[0] == 'b':
            num = num * 2
        new_name = "page_" + str(num) + ".jpg"
        os.rename(os.path.join(path,old_name),os.path.join(path,new_name))
        print("Finished page %d ..." % (num))
```

```python
path = os.getcwd()
rename_files(path)
```
